```python
# The purpose of this file is to transform the existing data CSV files into a format that will be usable for 
# analytics and Generative AI
import pandas as pd
```

```python
# Load FY22 data file and remove all records with a status of "Deleted"
data_file = pd.read_csv('./dsa-airflow/data/FY22Goals.csv', header=0, encoding = 'latin-1')

FY22_df= pd.DataFrame(data_file)

display(FY22_df.head(5))
```

```python
records_to_remove = ['Deleted']
FY22_df2 = FY22_df[~FY22_df['Active/Deleted'].isin(records_to_remove)]

display(FY22_df2.head(5))
```

```python
# drop the Status and Active/Deleted Column
FY22_df2.drop(columns=['Status','Active/Deleted'],inplace=True)

display(FY22_df2.head(10))
```

```python
#Add a FY column
FY22_df2["FY"]=[22]
display(FY22_df2.head(5))
```

```python

# Load FY23 data file and remove all records with a status of "Deleted"
data_file23 = pd.read_csv('./dsa-airflow/data/FY23Goals.csv', header=0, encoding = 'latin-1')

FY23_df= pd.DataFrame(data_file23)

display(FY23_df.head(5))
```

```python
records_to_remove = ['Deleted']
FY23_df2 = FY23_df[~FY23_df['Active/Deleted'].isin(records_to_remove)]

display(FY23_df2.head(5))
```

```python
# drop the Status and Active/Deleted Column
FY23_df2.drop(columns=['Status','Active/Deleted'],inplace=True)

display(FY23_df2.head(10))
```

```python
# Load FY24 data file 
data_file24 = pd.read_csv('./dsa-airflow/data/FY24Goals.csv', header=0, encoding = 'latin-1')

FY24_df= pd.DataFrame(data_file24)

display(FY24_df.head(5))
```

```python
# Combine all 3 dataframes into 1 data frame
Goals_df = pd.concat([FY22_df2,FY23_df2,FY24_df], axis=0)

display(Goals_df.head(5))

```
